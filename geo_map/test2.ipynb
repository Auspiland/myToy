```python
import geopandas as gpd
from shapely.geometry import Polygon
from shapely.validation import make_valid

# -------------------------------
# 설정
# -------------------------------

# ① shapefile 경로 (압축을 풀고 파일명을 여기에)
ADMIN_PATH = "gis_osm_admin_a_free_1.shp"

# ② 이름이 들어있는 컬럼명 (파일마다 다를 수 있음)
#    일반적으로 'name' 컬럼에 시군구 이름이 있습니다.
NAME_COL = "name"

# ③ 좌표계 (Geofabrik shapefile은 WGS84입니다)
SRC_CRS = "EPSG:4326"

# ④ 면적 계산용 좌표계 (한국 EPSG:5179 또는 5186)
AREA_CRS = "EPSG:5179"

# -------------------------------
# 함수 정의
# -------------------------------

def get_representative_region(rect_coords, admin_path=ADMIN_PATH, name_col=NAME_COL):
    """
    rect_coords: [(lon, lat), ...] 형태의 4개 점 (직사각형)
    """
    # 직사각형 polygon 생성
    xs, ys = zip(*rect_coords)
    minx, maxx = min(xs), max(xs)
    miny, maxy = min(ys), max(ys)
    rect_poly = Polygon([(minx, miny), (maxx, miny), (maxx, maxy), (minx, maxy)])

    # 행정구역 shapefile 로드
    gdf = gpd.read_file(admin_path)
    if gdf.crs is None:
        gdf.set_crs(SRC_CRS, inplace=True)
    gdf["geometry"] = gdf["geometry"].apply(make_valid)

    # 좌표계 변환 (면적 계산용)
    gdf = gdf.to_crs(AREA_CRS)
    rect_gdf = gpd.GeoDataFrame(geometry=[rect_poly], crs=SRC_CRS).to_crs(AREA_CRS)
    rect_poly_m = rect_gdf.geometry.iloc[0]

    # 교집합 면적 계산
    gdf["inter_area"] = gdf.geometry.intersection(rect_poly_m).area
    top = gdf.sort_values("inter_area", ascending=False).iloc[0]

    return {
        "name": top[name_col],
        "inter_area": float(top["inter_area"]),
        "total_area": float(rect_poly_m.area),
        "coverage_ratio": float(top["inter_area"] / rect_poly_m.area),
    }

# -------------------------------
# 사용 예시
# -------------------------------

if __name__ == "__main__":
    rect_coords = [
        (126.34827086975953, 38.449472749226594),
        (126.93263292526628, 38.453760025906824),
        (126.35549274508955, 37.98996965016792),
        (126.93619129881695, 37.994187205182236)
    ]

    result = get_representative_region(rect_coords)
    print(result)
```

```python
from collections import defaultdict

def find_duplicate_name_indexes(data):
    """
    GeoJSON 데이터에서 동일한 'name' 값을 가진 feature들의 인덱스를 반환.
    반환: {이름: [인덱스1, 인덱스2, ...]} 형식의 dict
    """
    name_to_idxs = defaultdict(list)

    for i, feature in enumerate(data.get("features", [])):
        props = feature.get("properties", {})
        name = props.get("name")
        if name:
            name_to_idxs[name].append(i)

    # 중복된 이름만 필터링
    return {k: v for k, v in name_to_idxs.items() if len(v) > 1}

find_duplicate_name_indexes(data)
```
